**IMPORT THE RELEVANT LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.datasets import load_breast_cancer
```

**READ IN THE DATASET**

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

**SCALE THE DATA**

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis=0)
```

**CREATE THE PRINCIPAL COMPONENT**

```python
pca_model = PCA(n_components=2)
```

```python
pc_result = pca_model.fit_transform(scaled_X)
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
cancer_dictionary.keys()
```

```python
cancer_dictionary['target']
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.scatter(pc_result[:,0], pc_result[:,1],c=cancer_dictionary['target'])
```

**ANALYSIS**

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1','PC2'], columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3), dpi=150)
sns.heatmap(df_comp, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
# from above the first pc explains 44.2% of the data and second pc explains 18.9% of the data
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
# if 2 pc explains over 60% of our data, what will happen if n_comp is = 30?
```

```python
pca_30 = PCA(n_components=30)
pc_30_result = pca_30.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.scatter(pc_30_result[:,0], pc_30_result[:,1],c=cancer_dictionary['target'])
```

```python
pca_30.components_
```

```python
df_30_comp = pd.DataFrame(pca_30.components_, index=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10',
                                                    'PC11','PC12','PC13','PC14','PC15','PC16','PC17','PC18','PC19','PC20',
                                                    'PC21','PC22','PC23','PC24','PC25','PC26','PC27','PC28','PC29','PC30'], 
                          columns=df.columns)
```

```python
df_30_comp
```

```python
plt.figure(figsize=(10,5), dpi=250)
sns.heatmap(df_30_comp, annot=True)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
        
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
# find out how many number of component is enough to explain the data
plt.figure(figsize=(10,5), dpi=250)
plt.plot(range(1,30),explained_variance)
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
```
