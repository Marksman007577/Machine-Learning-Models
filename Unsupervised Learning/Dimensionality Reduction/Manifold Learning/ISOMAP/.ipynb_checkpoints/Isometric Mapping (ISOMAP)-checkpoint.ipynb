**CREATE THE PRINCIPAL COMPONENT ANALYSIS (PCA)**

```python
from sklearn.decomposition import TruncatedSVD
```

```python
model = TruncatedSVD(n_components=200, algorithm='randomized', n_iter=5, random_state=101)
```

**FIT THE MODEL TO THE VARIOUS DATASET**

*train dataset*

```python
X_train_svd = model.fit_transform(X_train)
```

```python
X_train_svd = pd.DataFrame(data=X_train_svd, index=train_index)
```

```python
X_train_svd
```

*validation data*

```python
X_validation_svd = model.transform(X_validation)
```

```python
X_validation_svd = pd.DataFrame(data=X_validation_svd, index=validation_index)
```

```python
X_validation_svd
```

*test data*

```python
X_test_svd = model.transform(X_test)
```

```python
X_test_svd = pd.DataFrame(data=X_test_svd, index=test_index)
```

```python
X_test_svd
```

**EVALUATE THE VARIANCE OR INFORMATION OF THE NON-REDUCED DIMENSIONALITY**

*Because we have not reduced the dimensionality at all (we’ve just transformed the data) the variance/information of the original data captured by the 784 principal components should be 100%:*

```python
model_variance = model.explained_variance_ratio_
```

```python
print('Variance Explained by all 200 principal components:', sum(model_variance))
```

**OBTAIN THE IMPORTANCE OF VARIANCE FOR EACH SET OF PRINCIPAL COMPONENTS**

```python
importance_of_principal_components = pd.DataFrame(data=model_variance)
importance_of_principal_components = importance_of_principal_components.T
```

```python
importance_of_principal_components
```

```python
print('Variance Captured by First 10 Principal Components: ',importance_of_principal_components.loc[:,0:9].sum(axis=1).values)
print('Variance Captured by First 20 Principal Components: ',importance_of_principal_components.loc[:,0:19].sum(axis=1).values)
print('Variance Captured by First 50 Principal Components: ',importance_of_principal_components.loc[:,0:49].sum(axis=1).values)
print('Variance Captured by First 100 Principal Components: ',importance_of_principal_components.loc[:,0:99].sum(axis=1).values)
print('Variance Captured by First 200 Principal Components: ',importance_of_principal_components.loc[:,0:199].sum(axis=1).values)
```

*The first 10 components in total capture approximately 49% of the variance, the first one hundred components over 91%, and the first two hundred components over 96% of the variance; the information in the rest of the principal components is of negligible value.*

**VISUALIZE THE EFFECT OF THE COMPONENT SET VARIANCE FOR EACH DATASET**

```python
def scatter_plot(feature_dataframe, label_dataframe, algorithm_name):
    temp_dataframe = pd.DataFrame(data=feature_dataframe.loc[:, 0:1], index=feature_dataframe.index)
    temp_dataframe = pd.concat((temp_dataframe,label_dataframe), axis=1, join='inner')
    temp_dataframe.columns = ['First Vector', 'Second Vector', 'Label']
    sns.lmplot(x='First Vector', y='Second Vector', hue='Label', data= temp_dataframe, fit_reg=False)
    
    ax = plt.gca()
    ax.set_title(f'Seperation of Observations using {algorithm_name}')
```

*test data*

```python
scatter_plot(X_train_svd, y_train, 'Singular Value Decomposition (SVD)')
```

*validation data*

```python
scatter_plot(X_validation_svd, y_validation, 'Singular Value Decomposition (SVD)')
```

*test data*

```python
scatter_plot(X_test_svd, y_test, 'Singular Value Decomposition (SVD)')
```
