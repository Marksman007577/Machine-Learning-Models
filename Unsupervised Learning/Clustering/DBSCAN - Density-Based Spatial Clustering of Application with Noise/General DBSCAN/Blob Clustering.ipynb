**COMPARING DBSCAN AND K-MEANS CLUSTERING**

**IMPORT RELEVANT LIBRARIES**

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1',y='X2', hue=labels, palette='Set1')
    
```

```python
KMeans_Model = KMeans(n_clusters=3)
display_categories(model=KMeans_Model, data=blobs)
```

```python
KMeans_Model = KMeans(n_clusters=2)
display_categories(model=KMeans_Model, data=moons)
```

```python
KMeans_Model = KMeans(n_clusters=2)
display_categories(model=KMeans_Model, data=circles)
```

```python
DBSCAN_Model = DBSCAN()
display_categories(model=DBSCAN_Model, data=blobs)
```

```python
DBSCAN_Model = DBSCAN(eps=0.15)
display_categories(model=DBSCAN_Model, data=moons)
```

```python
DBSCAN_Model = DBSCAN(eps=0.15)
display_categories(model=DBSCAN_Model, data=circles)
```

**IMPLEMENTING DBSCAN WITH HYPERPARAMETER SEARCH**

```python
two_blobs = pd.read_csv('cluster_two_blobs.csv')
```

```python
two_blobs 
```

```python
sns.scatterplot(data=two_blobs, x='X1',y='X2')
```

```python
two_blobs_outliers = pd.read_csv('cluster_two_blobs_outliers.csv')
```

```python
two_blobs_outliers
```

```python
sns.scatterplot(data=two_blobs_outliers, x='X1',y='X2')
```

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1',y='X2', hue=labels, palette='Set1')
```

```python
DBSCAN_Model = DBSCAN() 
```

```python
display_categories(model=DBSCAN_Model, data=two_blobs)
```

```python
display_categories(model=DBSCAN_Model, data=two_blobs_outliers)
```

```python
DBSCAN_Model_Parameter = DBSCAN(eps=1.5)
display_categories(model=DBSCAN_Model_Parameter, data=two_blobs_outliers)
```

```python
DBSCAN_Model_Parameter.labels_
```

```python
# total outliers found
np.sum(DBSCAN_Model_Parameter.labels_ == -1)
```

```python
# percentage of points classified as outliers
100 * np.sum(DBSCAN_Model_Parameter.labels_ == -1) / len(DBSCAN_Model_Parameter.labels_)
```

```python
len(DBSCAN_Model_Parameter.labels_)
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001,7,200):
    DBSCAN_Model_Parameter = DBSCAN(eps=eps)
    DBSCAN_Model_Parameter.fit(two_blobs_outliers)
    
    # total outliers found
    number_of_outliers.append(np.sum(DBSCAN_Model_Parameter.labels_ == -1))
    
    # percentage of points classified as outliers
    perc_outliers = 100 * np.sum(DBSCAN_Model_Parameter.labels_ == -1) / len(DBSCAN_Model_Parameter.labels_)
    
    outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.linspace(0.001,7,200), y=number_of_outliers)
plt.xlim(0,2)
plt.ylim(0,10)
plt.ylabel('Percentage of Points Classified as Outliers')
plt.hlines(y=3, xmin=0, xmax=2, color='red')
```

```python
outlier_percent = []
number_of_outliers = []

for n in np.arange(1,100):
    DBSCAN_Model_Parameter = DBSCAN(min_samples=n)
    DBSCAN_Model_Parameter.fit(two_blobs_outliers)
    
    # total outliers found
    number_of_outliers.append(np.sum(DBSCAN_Model_Parameter.labels_ == -1))
    
    # percentage of points classified as outliers
    perc_outliers = 100 * np.sum(DBSCAN_Model_Parameter.labels_ == -1) / len(DBSCAN_Model_Parameter.labels_)
    
    outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.arange(1,100), y=outlier_percent)
plt.xlabel('Minimum Number of Samples')
plt.ylabel('Percentage of Points Classified as Outliers')
```

**FINAL MODEL**

```python
num_dim = two_blobs_outliers.shape[1]
DBSCAN_Model_Final = DBSCAN(min_samples=2*num_dim, eps=0.75)
display_categories(model=DBSCAN_Model_Final, data=two_blobs_outliers)
```
