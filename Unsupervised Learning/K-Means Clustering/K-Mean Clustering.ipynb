**CREATE THE FEATURES WITH ALL CATEGORICAL COLUMNS SET WITH DUMMY VALUES**

```python
X = pd.get_dummies(df)
```

```python
X
```

**SCALE THE DATA**

```python
# create an instance of the scaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

**CREATE THE CLUSTER MODEL**

```python
KMeans_Model = KMeans(n_clusters=2)
```

**FIT AND PREDICT THE MODEL**

```python
cluster_labels = KMeans_Model.fit_predict(scaled_X)
```

```python
cluster_labels
```

**MERGE ORIGINAL DATA WITH THE CLUSTER**

```python
X['Clusters'] = cluster_labels
```

```python
X
```

```python
X.corr()
```

```python
plt.figure(figsize=(12,5), dpi=300)
X.corr()['Clusters'].iloc[:-1].sort_values().plot(kind='bar')
```

```python
ssd = []

for k in range(2,10):
    KMeans_Model = KMeans(n_clusters=k)
    KMeans_Model.fit(scaled_X)
    
    
    ssd.append(KMeans_Model.inertia_) # SSD points --> cluster center
    
```

```python
ssd
```

```python
plt.plot(range(2,10), ssd, 'o--')
```
