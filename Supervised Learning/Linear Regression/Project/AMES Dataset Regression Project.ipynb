```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import scipy as sp
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error
from joblib import  dump, load

def root_mean_square_error(label_test, test_predictions):
    """function to perform root mean squared error"""
    rmse = mean_squared_error(label_test, test_predictions)
    return math.sqrt(rmse)


# 1. importing dataset
df = pd.read_csv('AMES_Final_DF.csv')
df.head()
df.info()

# 2. The label we are trying to predict is the SalePrice column
# Seperate out the data into X features and  y label

X = df.drop("SalePrice", axis=1) # drops the label called sales

#b. creating the label
y = df['SalePrice']

# 3. Performing the train test split on poly features
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)

# 4 performing scaling
#a. initializing the standard scaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)

#b. obtain X_test
scaled_X_test = scaler.transform(X_test)

# 5. Obtain base elastic net and create an instance
base_elastic_net_model = ElasticNet(max_iter=1000)
param_grid = {'alpha':[0.1,1,5,10,50,100],'l1_ratio':[.1,.5,.7,.95,.99,1]}

# 6. creating the model
grid_model = GridSearchCV(estimator=base_elastic_net_model, param_grid=param_grid, scoring='neg_mean_squared_error',cv=5, verbose=2)

# 7 fit the prefered model on the whole data set
grid_model.fit(scaled_X_train, y_train)

grid_model.best_estimator_

# 8. Get the prediction value from the test data 
y_prediction = grid_model.predict(scaled_X_test)

# 11. Evaluate error using MSE
MAE = mean_absolute_error(y_test, y_prediction)
MSE = mean_squared_error(y_test, y_prediction)
RMSE = root_mean_square_error(label_test=y_test, test_predictions=y_prediction)
```
