```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import scipy as sp

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.pairplot(data=df)
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30, random_state=101)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_bar = model.predict(X_test)
```

```python
y_bar
```

```python
error = y_test-y_bar
```

```python
error
```

```python
MAE = mean_absolute_error(y_test, y_bar)
```

```python
MAE
```

```python
MSE = mean_squared_error(y_test, y_bar)
```

```python
MSE
```

```python
RMSE = math.sqrt(MSE)
```

```python
RMSE
```

```python
plt.figure(figsize=(10,5), dpi=150)
sns.scatterplot(x=y_test, y=error)
plt.axhline(y=0, color='r', ls='--')
```

```python
# plotting the distribution
plt.figure(figsize=(10,5), dpi=200)
sns.displot(error, bins=20, kde=True)
```

```python
# Normal probability plot
# Helps to compare what a normal distribution plot (red line) would look like against your plot/values
fig, ax = plt.subplots(figsize=(6,5), dpi=200)
_ = sp.stats.probplot(error, plot=ax)
```

```python
model.coef_
```

```python
# Visualizing the prediction against the true features value
y_bar_allfeatures = model.predict(X)
```

```python
y_bar_allfeatures
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].plot(df['TV'], y_bar_allfeatures, 'o', color='red')
axes[0].set_ylabel('Sales')
axes[0].set_title('TV Spend')

axes[1].plot(df['radio'], df['sales'], 'o')
axes[1].plot(df['radio'], y_bar_allfeatures, 'o', color='red')
axes[1].set_title('Radio Spend')
axes[1].set_ylabel('Sales')

axes[2].plot(df['newspaper'], df['sales'], 'o')
axes[2].plot(df['newspaper'], y_bar_allfeatures, 'o', color='red')
axes[2].set_title('Newspaper Spend');
axes[2].set_ylabel('Sales')


plt.tight_layout();
```
