```python
# importing the necessary libraries
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.metrics import plot_precision_recall_curve,plot_roc_curve,roc_curve, auc

from sklearn.pipeline import Pipeline
```

```python
# reading the dataset
df = pd.read_csv("sonar.csv")
```

```python
df
```

```python
df.head(5)
```

Exploratory data analysis

```python
# generate the heat map
plt.figure(figsize=(12,8), dpi=200)
sns.heatmap(df.corr(), cmap='coolwarm')
```

CONVERT THE LABEL VALUES TO A DIGIT 0 OR 1

```python
df['Target'] = df['Label'].map({'R':0, 'M':1})
```

```python
df
```

```python
# creating the features and label
X = df.drop(['Label','Target'], axis=1)
y = df['Target']
```

```python
X
```

```python
y
```

```python
# perform train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,random_state = 42)
```

```python
# scaling the data
scaler = StandardScaler()
```

```python
# creating the model using KNeighborClassifier
knn_model = KNeighborsClassifier()
```

```python
# PIPELINE --> GRIDSEARCH
```

```python
operations = [('scaler',scaler),('knn',knn_model)]
```

```python
pipe = Pipeline(operations)
```

```python
k_values = list(range(1,30))
```

```python
k_values
```

```python
param_grid = {'knn__n_neighbors':k_values}
```

```python
# setup grid search
full_cv_classifier = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train,y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
# using the .cv_results_dictionary to create a plot of the mean test score per K value
full_cv_result = full_cv_classifier.cv_results_
```

```python
full_cv_result
```

```python
full_cv_result_dataframe = pd.DataFrame(full_cv_result)
```

```python
full_cv_result_dataframe
```

```python
full_cv_result_dataframe.mean_test_score
```

```python
full_cv_result_dataframe.mean_test_score.plot()
```

```python
full_pred = full_cv_classifier.predict(X_test)
```

```python
full_pred
```

```python
# evaluating the model performance
accuracy_score(y_test, full_pred)
```

```python
# error
1 - accuracy_score(y_test, full_pred)
```

```python
# evaluate confusion matrix
confusion_matrix(y_test, full_pred)
```

```python
# print classification report
print(classification_report(y_test,full_pred))
```
