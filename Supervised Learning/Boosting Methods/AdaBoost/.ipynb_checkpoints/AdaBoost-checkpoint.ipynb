DATA HISTORY

Mushroom Hunting: Edible or Poisonous?
Data Source: https://archive.ics.uci.edu/ml/datasets/Mushroom

This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.

Goal
THIS IS IMPORTANT, THIS IS NOT OUR TYPICAL PREDICTIVE MODEL!

Our general goal here is to see if we can harness the power of machine learning and boosting to help create not just a predictive model, but a general guideline for features people should look out for when picking mushrooms.

IMPORT RELEVANT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import plot_confusion_matrix, classification_report, confusion_matrix, accuracy_score
```

READ IN THE DATASET

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df
```

PERFORM EXPLORATORY DATA ANALYSIS

```python
sns.countplot(data=df, x='class')
```

```python
df.describe()
```

```python
unique_features = df.describe().transpose().reset_index().sort_values('unique')
```

```python
unique_features
```

```python
plt.figure(figsize=(14,6), dpi=200)
plt.xticks(rotation=90)
sns.barplot(data=unique_features, x='index', y='unique')
```

PERFROM FEATURE AND LABEL SPLIT

```python
# checking for missing values
df.isnull().sum()
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
# use dummy to convert categorical data to numerical
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

PERFORM TRAIN TEST SPLIT

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15, random_state=101)
```

CREATE THE ADABOOST MODEL

```python
Base_AdaBoost_Model = AdaBoostClassifier(n_estimators=1)
```

```python
# Fit the model to the training model
Base_AdaBoost_Model.fit(X_train,y_train)
```

```python
# predict on the test dataset using the model
Base_AdaBoost_Model_Prediction = Base_AdaBoost_Model.predict(X_test)
```

```python
Base_AdaBoost_Model_Prediction
```

```python
# plot the confusion matirx
plot_confusion_matrix(Base_AdaBoost_Model,X_test,y_test)
```

```python
# compare the prediction to the truth label
print(classification_report(y_test, Base_AdaBoost_Model_Prediction))
```

```python
# view the relevance of each feature importance
Base_AdaBoost_Model.feature_importances_
```
