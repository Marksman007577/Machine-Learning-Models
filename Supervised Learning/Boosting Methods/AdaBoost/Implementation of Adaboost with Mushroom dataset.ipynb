**IMPORT RELEVANT LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**READ THE DATASET**

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df
```

**PERFORM EXPLORATORY DATA ANALYSIS**

```python
sns.countplot(data=df, x='class')
```

**PERFORM FEATURE AND LABEL SPLIT**

```python
X = df.drop(columns='class', axis=1)
X = pd.get_dummies(X, drop_first=True, dtype='uint8')
```

```python
X
```

```python
y = df['class']
```

```python
y
```

**PERFORM TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**CREATE THE ADABOOST MODEL**

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

```python
model
```

**FIT MODEL TO TRAIN FEATURE AND LABEL DATASET**

```python
model.fit(X_train, y_train)
```

**PREDICT USING THE MODEL ON THE TEST FEATURE DATASET**

```python
pred = model.predict(X_test)
```

```python
pred
```

**OBTAIN THE ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
```

```python
acc_score = accuracy_score(y_test, pred)
```

```python
acc_score
```

**OBTAIN THE CONFUSION MATRIX, CONFUSION MATRIX DISPLAY, AND CLASSIFICATION REPORT**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
confusion_matrix(y_test, pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

```python
model.feature_importances_
```

**OBTAIN THE BEST ESTIMATOR BY RUNNING DIFFERENT ESTIMATORS**

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

error_rates = []

for n in range(1, 96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc_score = accuracy_score(y_test, pred)
    error = 1 - acc_score
    error_rates.append(error)
    

model_feature_importance = model.feature_importances_
feature_dataframe = pd.DataFrame(index=X.columns, data=model_feature_importance, columns=['Importance'])
important_features = feature_dataframe[feature_dataframe['Importance']>0]
```

```python
error_rates
```

```python
plt.figure(figsize=(14,6), dpi=200)
plt.plot(range(1,96), error_rates)
```

```python
important_features
```

```python
model_feature_importance
```

```python
plt.figure(figsize=(14,6), dpi=200)
plt.xticks(rotation=90)
sns.barplot(data=important_features.sort_values('Importance'),x=important_features.index, y='Importance')
```

**CREATE THE FINAL MODEL**

```python
final_model = AdaBoostClassifier(n_estimators=10)
```

```python
final_model.fit(X_train, y_train)
```

```python
final_predict = final_model.predict(X)
```

```python
final_predict
```

```python
accuracy_score(y, final_predict)
```

```python
confusion_matrix(y, final_predict)
```

```python
ConfusionMatrixDisplay.from_predictions(y, final_predict)
```

```python
print(classification_report(y, final_predict))
```
