**IMPORT RELEVANT LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**READ THE DATASET**

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df
```

**PERFORM EXPLORATORY DATA ANALYSIS**

```python
sns.countplot(data=df, x='class')
```

```python
df.info()
```

```python
unique_feature = df.describe().transpose().reset_index().sort_values('unique')
```

```python
unique_feature
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(14,6), dpi=200)
plt.xticks(rotation=90)
sns.barplot(data=unique_feature,x='index', y='unique')
```

**PERFORM FEATURE AND LABEL SPLIT**

```python
X = df.drop(columns='class', axis=1)
X = pd.get_dummies(X, drop_first=True, dtype='uint8')
```

```python
X
```

```python
y = df['class']
```

```python
y
```

**PERFORM TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**CREATE THE GRADIENT BOOST MODEL INSTANCE**

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_model = GradientBoostingClassifier()
```

**CREATE THE GRIDSEARCHCV INSTANCE**

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50,100],
             'learning_rate':[0.1,0.05,0.2],
             'max_depth':[3,4,5]}
```

```python
grid_model = GridSearchCV(estimator=gb_model, param_grid=param_grid)
```

```python
grid_model
```

**FIT GRID MODEL TO TRAIN FEATURE AND LABEL DATASET**

```python
grid_model.fit(X_train, y_train)
```

**OBTAIN THE BEST ESTIMATOR**

```python
grid_model.best_estimator_
```

**PREDICT USING THE GRID MODEL ON THE TEST FEATURE DATASET**

```python
grid_model_pred = grid_model.predict(X_test)
```

```python
grid_model_pred
```

**OBTAIN THE BEST PARAMETERS FROM THE GRIDSEARCHCV**

```python
grid_model.best_params_
```

**OBTAIN THE ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
```

```python
acc_score = accuracy_score(y_test, grid_model_pred)
```

```python
acc_score
```

**OBTAIN THE CONFUSION MATRIX, CONFUSION MATRIX DISPLAY, AND CLASSIFICATION REPORT**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
confusion_matrix(y_test, grid_model_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, grid_model_pred)
```

```python
print(classification_report(y_test, grid_model_pred))
```

**OBTAIN THE DIFFERENT FEATURE IMPORTANCE**

```python
model_estimator = grid_model.best_estimator_
```

```python
model_estimator
```

```python
model_feature_importance = model_estimator.feature_importances_
```

```python
model_feature_importance
```

```python
feature_dataframe = pd.DataFrame(index=X.columns, data=model_feature_importance, columns=['Importance'])
```

```python
feature_dataframe
```

```python
important_features = feature_dataframe[feature_dataframe['Importance']>0]
```

```python
important_features
```

```python
plt.figure(figsize=(14,6), dpi=200)
plt.xticks(rotation=90)
sns.barplot(data=important_features.sort_values('Importance'),x=important_features.index, y='Importance')
```

**CREATE FINAL MODEL**

```python
final_model = GradientBoostingClassifier(learning_rate=0.1, max_depth=3, n_estimators=50)
```

```python
final_model
```

```python
final_model.fit(X_train, y_train)
```

```python
final_pred = final_model.predict(X)
```

```python
final_pred
```

```python
confusion_matrix(y,final_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y, final_pred)
```

```python
print(classification_report(y, final_pred))
```
