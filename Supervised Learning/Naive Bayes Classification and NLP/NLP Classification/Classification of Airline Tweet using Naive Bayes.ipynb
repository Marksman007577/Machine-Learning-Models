**EXTRACT THE DATA FOR ANALYSIS**

```python
data = df[['text','airline_sentiment']]
```

```python
data
```

**CREATE FEATURE AND LABEL**

```python
X = data['text']
```

```python
X
```

```python
y = data['airline_sentiment']
```

```python
y
```

**PERFROM TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**PERFROM VECTORIZATION FOR FITTING AND TRANSFORMING THE TRAIN AND TEST FEATURES**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid_vect = TfidfVectorizer(stop_words='english')
```

**FIT AND TRANSFORM THE FEATURE DATA SET**

```python
X_train_vector = tfid_vect.fit_transform(X_train)
```

```python
X_test_vector = tfid_vect.transform(X_test)
```

**BUILD/CREATE THE MODEL**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb_model = MultinomialNB(alpha=0.1)
```

**FIT THE MODEL ON THE VECTORIZED TRAIN FEATURE AND LABEL**

```python
nb_model.fit(X_train_vector, y_train)
```

**PREDICT ON THE TEST FEATURE DATASET**

```python
pred = nb_model.predict(X_test_vector)
```

**OBTAIN THE CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test, pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, pred)
```

**OBTAIN THE ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

**OBTAIN CLASSIFICATION REPORT**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

**CREATE A PIPELINE FOR THE FINAL MODEL**

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline(steps=[('tfid', TfidfVectorizer()),
                           ('naive_baye', MultinomialNB())]
                   )
```

```python
pipeline.fit(X, y)
```

```python
pipeline.predict(['air fright'])
```
