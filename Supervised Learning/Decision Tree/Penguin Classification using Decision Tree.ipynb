**PERFROM FEATURE AND LABEL SPLIT**

```python
X = df.drop('species', axis=1)
```

```python
y = df['species']
```

```python
X
```

```python
df['island'].unique()
```

```python
y
```

**CONVERT CATEGORICAL DATA TO NUMERICAL DATA**

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

**PERFORM TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**CREATE AN INSTANCE OF A DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

**FIT MODEL TO THE TRAIN FEATURE AND LABEL**

```python
model.fit(X_train, y_train)
```

**PREDICT MODEL ON TEST FEATURE**

```python
pred = model.predict(X_test)
```

```python
pred
```

**OBTAIN THE CONFUSION MATRIX, CONFUSION MATRIX DISPLAY, AND CLASSIFICATION REPORT**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
confusion_matrix(y_test, pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

**OBTAIN ACCURACY SCORE, AND ERROR**

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

**OBTAIN THE RELEVANCE OF EACH FEATURE IMPORTANCE**

```python
model.feature_importances_
```

```python
pd.DataFrame(data=model.feature_importances_, index=X.columns, columns=['Feature Importance']).sort_values('Feature Importance')
```

**VISUALIZE THE DECISION TREE**

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8), dpi=200)
plot_tree(model, filled=True);
```

**CREATING A FUNCTION TO REPORT THE MODEL IN ONE SHOT**

```python
from sklearn.tree import plot_tree

def report_model(a_model):
    a_model.fit(X_train, y_train)
    pred = a_model.predict(X_test)
    
    print(classification_report(y_test, pred))
    print('\n')
    
    plt.figure(figsize=(12,8), dpi=200)
    plot_tree(a_model, feature_names=X.columns, filled=True);
```

```python
tree_model = DecisionTreeClassifier(max_depth=5, criterion='entropy', max_leaf_nodes=None)
```

```python
tree_model
```

```python
report_model(a_model=tree_model)
```
